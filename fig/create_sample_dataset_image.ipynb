```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
left = imread("left.jpg")
center = imread("center.jpg")
right = imread("right.jpg")
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(14,3))

ax1.imshow(left)
ax1.set_axis_off()
ax1.set_title("left")

ax2.imshow(center)
ax2.set_axis_off()
ax2.set_title("center")

ax3.imshow(right)
ax3.set_axis_off()
ax3.set_title("right")

plt.tight_layout()
fig.savefig("sample_dataset.png")
```
